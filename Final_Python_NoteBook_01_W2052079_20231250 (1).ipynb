Tutorial 01 - Python Lab Seminar 02- Part (D)

```python
# --- Step 1: Define IQR outlier removal function with tracking ---
def drop_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    original_count = df.shape[0]
    df_cleaned = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    removed_count = original_count - df_cleaned.shape[0]
    print(f"{col}: Removed {removed_count} outliers (Remaining: {df_cleaned.shape[0]})")
    return df_cleaned

# --- Step 2: Visualize before cleaning ---
fig_age = px.box(df, y='Age', title='Age - Before Outlier Removal')
fig_age.show()

fig_tumor = px.box(df, y='Tumor_Size', title='Tumor Size - Before Outlier Removal')
fig_tumor.show()

fig_survival = px.box(df, y='Survival_Months', title='Survival Months - Before Outlier Removal')
fig_survival.show()

fig_nodes_examined = px.box(df, y='Regional_Node_Examined', title='Regional Nodes Examined - Before Outlier Removal')
fig_nodes_examined.show()

fig_nodes_positive = px.box(df, y='Regional_Node_Positive', title='Regional Positive - Before Outlier Removal')
fig_nodes_positive.show()

# --- Step 3: Drop outliers using IQR ---
print("\n--- Outlier Removal Report ---")
for col in ['Age', 'Tumor_Size', 'Survival_Months']:
    df = drop_outliers_iqr(df, col)

# --- Step 4: Apply logical validation ---
before_logic = df.shape[0]
df = df[(df['Regional_Node_Positive'] >= 0) &
        (df['Regional_Node_Examined'] >= 0) &
        (df['Regional_Node_Positive'] <= df['Regional_Node_Examined'])]
after_logic = df.shape[0]
print(f"Logical validation: Removed {before_logic - after_logic} rows (Remaining: {after_logic})")

# --- Step 5: Visualize after cleaning ---
fig_age_clean = px.box(df, y='Age', title='Age - After Outlier Removal')
fig_age_clean.show()

fig_tumor_clean = px.box(df, y='Tumor_Size', title='Tumor Size - After Outlier Removal')
fig_tumor_clean.show()

fig_survival_clean = px.box(df, y='Survival_Months', title='Survival Months - After Outlier Removal')
fig_survival_clean.show()

fig_nodes_examined_clean = px.box(df, y='Regional_Node_Examined', title='Regional Nodes Examined - After Outlier Removal')
fig_nodes_examined_clean.show()

fig_nodes_positive_clean = px.box(df, y='Regional_Node_Positive', title='Regional Positive - After Outlier Removal')
fig_nodes_positive_clean.show()

# --- Final row count ---
print(f"\nFinal dataset size: {df.shape[0]} rows")
```
